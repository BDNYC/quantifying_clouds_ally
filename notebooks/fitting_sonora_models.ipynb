In the diamondback-data folder we have included Sonora Diamondback models (Morley et al. 2024) with temperatures 1300–2400 K (notated by the folders in diamondback-data), surface gravity in the range log g = 3.5 − 5.5, and cloud parameter (fsed) from 1 (thick clouds) to 8 (thin
clouds) as well as cloud-free cases. The metallicity of all of these models is m =0.0, which is solar levels.

We convolved the Diamondback models to a resolution of R = 1000 at the 1.17 μm potassium doublet, corresponding
to the JWST NIRSpec medium resolution. Subsequently, we fit the pseudo-Voigt to each model using Doublet_Quantifier.curvefit.fit_two_curves. Further details on this function is in functions/data_tools.py

```python
# getting files of data
sonora_temps = [900, 1000, 1100, 1200, 1300, 1400, 1500, 1600, 1700, 1800, 1900, 2000, 2100, 2200, 2300, 2400]

# Note: running this code will take a long time, so it is recommend to only run it once to produce the pickle files
create_pickle_files = False
# if you wish to see the indiviual fits plottend append the temperatures you wish to see to the list below
show_pseudo_voigt = [2000]


for temp in sonora_temps:
    temp_path = mypath + '/diamondback-data/t' + str(temp) + '/' 
    files = [f for f in os.listdir(temp_path)]
    
    if '.DS_Store' in files: 
        # .DS_Store file is a hidden file on a Mac that stores information about a folder's custom attributes
        # we don't want to include it in our list of files
        files.remove('.DS_Store')

    if temp in show_pseudo_voigt or create_pickle_files:
        print("Effective Temperature: ", temp)
        show_plots = temp in show_pseudo_voigt
        # data_dict is a dictionary with the file names as the keys and the data as the values
        data_dict = files_to_spec(files, path_to_files  = temp_path) 
        # this is the step where we convert fit the pseudo-voigt function to the data
        parameter_df, convolve_data_dict = spec_to_parameter(data_dict, resolution= 1000, show_plots= show_plots)

    if create_pickle_files:
        # pickling the data produced above
        parameter_df.to_pickle(
            mypath + "/diamondback-data/pickle_files/parameters/" + temp + "_parameter_df.pkl")

        with open(mypath+'/diamondback-data/pickle_files/convolved_data/' + temp +'_convolve_data_dict.pkl', 'wb') as handle:
            pickle.dump(convolve_data_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
